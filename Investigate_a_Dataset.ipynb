### 1. Which country has the highest number of cell phone owners in 2020?

```python
#so let us plot the data in that specific year
# Since we have a lot of data (211,42) we can't plot that large data in a single plot. So, I have segmented the data into
#three segments; starting from row 1 upto row 70 is segment 1, segment 2 starts from 71 up to 140 and segment 3 is from 141-211

#df1 = df.iloc[:70,:]
#df2 = df.iloc[71:140,:]
#df3 = df.iloc[141:211,:]
```

```python
#in here, I have used a function to combine the different segments, and then I will call the function by making the three
#different segments as an input
def func(df1):
    axs = plt.subplots(1, figsize=(20, 10))
    saw = sns.barplot(x = "Country Name", y = '2020', data = df1, ax=axs[1])
    saw.set_xticklabels(saw.get_xticklabels(), 
                          rotation=90, 
                          horizontalalignment='right');
    return saw
```

```python
#let us call the function where the input is the first segment
func(df.iloc[:70,:])
```

```python
func(df.iloc[71:140,:])
```

```python
#this bar chart shows the third segment in our dataset
func(df.iloc[141:211,:])
```

In order to answer the question, we have to compare these three bar charts, then we have specify the country which has the highest number of cell phone owners in 2020. That is China, it had a large number of mobile cell phones in 2020. 

### 2. In which year did the number of cell phones that have been using reach in 1 billion?

```python
#to answer this question, I have take sample of the dataset, for example, the first 25 columns excluding country column 
#contain the year which the number of cell phones reached 1 billion in number through out the world

#so this code below chooses the first 25 columns excluding the country column.
w = df.iloc[:,1:25].sum(axis=0)

#in here, each column contains the total number of cell phones in every country with in specific year. let us look at it
w

#when you run the code, you will first see that i.e, in 1980, the total number of cell phones that have been used through out
#the world is 23482
```

```python
#in order to get the first year that the total number of phones reached in 1 million, we have visualize our dataset by using
#line plot, so this is the result
plt.figure(figsize = (23,12));
gr = plt.plot(w,'ro-',linewidth=3)
plt.grid()
```

So, the answer of the above question is 2002, this is the year that the total number of cell phones through out the world reachd 1 Billion in number.

### 3. In 2000, the world population was estimated 6.143 billion by UN world population prospects <a href="https://www.worldometers.info/world-population/world-population-by-year/">(refer),</a> So, what percentage of the world population has access to cell phones in 2000? 

```python
#referrint to the above figure which tells total number of cell phones per year, we can say that the number of peaple those had
#access to cell phonse in 2000 was approximately 700 million.

w = df.iloc[:,20:23].sum(axis=0)
plt.figure(figsize = (20,10));
gr = plt.plot(w,'ro-',linewidth=3)
plt.grid()
```

```python
#in order to know the exact value, we can use sum function
df['2000'].sum()

#the percentage of the world population that had access to cell phones in 2000 can be found by this code.

percent = df['2000'].sum()/6143000000*100
percent

#so, 11.7% of the world population at that time (2000) had access to cell phones
```

### 4. Which country was having the highest number of cell phones in  end of last century?
